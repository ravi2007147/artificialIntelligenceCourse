```python
#7.print prime numbers between 100 to 200.
def get_prime_numbers(start,end):
    primelist=[]

    for x in (range(start,end+1)):
        number = x
        is_prime=True
        for j in (range(2,int(number**0.5)+1)):
            if number % j == 0:
               is_prime=False
               break

        if is_prime:
            primelist.append(number)
    return primelist

print("Prime numbers are ",', '.join(str(num) for num in get_prime_numbers(100,200)))
```

```python
#13. Print fibonacci series till 50.
def print_fib(number):
    fiblist = [0,1]
    loop = True
    while loop:
        fiblist.append(fiblist[-2]+fiblist[-1])

        if (fiblist[-2]+fiblist[-1]) > number:
            loop = False

    return fiblist

print(print_fib(50))
```

```python
# 1. Check how many charectors are repeating -"Elephent", "Icecream"
def find_repeating_characters(characters):
    new_list=list(set(characters.lower()))
    total_repeating=0
    for x in range(0,len(new_list)):
        if(characters.lower().count(new_list[x])>1):
            total_repeating = total_repeating + 1

    print("Total repeating characters are ",total_repeating)

find_repeating_characters("Elephent")
```

```python
#9.print duplicates from the list.
def print_duplicates(lst):
    seen = {}
    duplicates = []

    for item in lst:
        if item in seen:
            if item not in duplicates:
                duplicates.append(item)
        else:
            seen[item] = True

    print("Duplicates:", duplicates)

my_list = [10, 20, 30, 20, 20, 30, 40, 50, -20, 60, 60, -20, -20]
print_duplicates(my_list)
```

```python
#print binomial number till 80.
# Function to calculate binomial coefficient
def binomial_coefficient(n, k):
    # Calculate binomial coefficient using dynamic programming
    dp = [[0] * (k+1) for _ in range(n+1)]
    for i in range(n+1):
        for j in range(min(i, k)+1):
            if j == 0 or j == i:
                dp[i][j] = 1
            else:
                dp[i][j] = dp[i-1][j-1] + dp[i-1][j]
    return dp[n][k]


def print_binomial_numbers(up_to):
    for n in range(up_to + 1):
        if n > 80:
            break
        for k in range(n + 1):
            binomial = binomial_coefficient(n, k)
            print(f"C({n}, {k}) = {binomial}")

print_binomial_numbers(80)
```

10.Python program to create a list of tuples from given list having number and its cube in each tuple.
Example:

Input: list = [1, 2, 3]
Output: [(1, 1), (2, 8), (3, 27)]

```python
def create_tuple_list(lst):
    tuple_list = [(num, num**3) for num in lst]
    return tuple_list

input_list = [1, 2, 3]
output_list = create_tuple_list(input_list)
print(output_list)
```
