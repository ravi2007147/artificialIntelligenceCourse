```python
df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                   "B":[None, 2, 54, 3, None],
                   "C":[20, 16, None, 3, 8],
                   "D":[14, 3, None, None, 6]})

print("Fill with specific value")
print(df.fillna(0))
print("\n")

df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                   "B":[None, 2, 54, 3, None],
                   "C":[20, 16, None, 3, 8],
                   "D":[14, 3, None, None, 6]})
print("Drop Empty value")
print(df.dropna())
print("\n")

df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                   "B":[None, 2, 54, 3, None],
                   "C":[20, 16, None, 3, 8],
                   "D":[14, 3, None, None, 6]})
print("It will start from the top and fill the value from the previous value")
print(df.ffill())
print("\n");

df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                   "B":[None, 2, 54, 3, None],
                   "C":[20, 16, None, 3, 8],
                   "D":[14, 3, None, None, 6]})

print("It will start from the bottom and fill the value from the previous value")
print(df.bfill())
print("\n");

df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                   "B":[None, 2, 54, 3, None],
                   "C":[20, 16, None, 3, 8],
                   "D":[14, 3, None, None, 6]})

print("It will start from the end and fill the value from the previous value, it will fill across rows")
print(df.bfill(axis="columns"))
print("\n");

df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                   "B":[None, 2, 54, 3, None],
                   "C":[20, 16, None, 3, 8],
                   "D":[14, 3, None, None, 6]})
import numpy as np
print("Replace a value")
print(df.replace({np.nan:30}))
print("\n");

import numpy as np
print("Interpolate method which replace based on the another values")
print(df.interpolate())
print("\n");
```
