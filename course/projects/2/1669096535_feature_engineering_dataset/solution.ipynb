```python
#no highly corelated data
def getHighCofficient(values_list,dfLocal):
    matches = []

    for i, value1 in enumerate(values_list):
        for j, value2 in enumerate(values_list):
            if value1 and value2:
                try:
                    if(dfLocal[value1].corr(dfLocal[value2])>=0.9 and value1!=value2):
                        matches.append([value1, value2,dfLocal[value1].corr(dfLocal[value2])])
                except:
                    pass
    return matches
                
getHighCofficient(subset_df.columns.tolist(),subset_df)
```

```python
from scipy.stats import f_oneway
# Perform one-hot encoding
one_hot_encoded = pd.get_dummies(df['Alley'], prefix='Alley')
# Concatenate the one-hot encoded columns with the original DataFrame
df_encoded = pd.concat([df, one_hot_encoded], axis=1)
# Visualization: Box plot
sns.boxplot(data=df_encoded, x='Alley', y='SalePrice')
```

```python
# ANOVA Test
# print(one_hot_encoded.columns)
one_hot_encoded_columns = one_hot_encoded.columns
category_groups = [df_encoded[df_encoded[column] == 1]['SalePrice'] for column in one_hot_encoded_columns]
f_statistic, p_value = f_oneway(*category_groups)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```
