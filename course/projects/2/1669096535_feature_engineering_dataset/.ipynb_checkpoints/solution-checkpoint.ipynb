```python
#dropping columns with missing lot of data
df=df.drop(['PoolQC','MiscFeature','Alley','Fence'],axis=1)
```

```python
#FireplaceQu also has high missing values but this variable has some meaning and might be affecting the SalePrice
#Fireplace has the following categories:

#Ex Excellent – Exceptional Masonry Fireplace
#Gd Good – Masonry Fireplace in main level
#TA Average – Prefabricated Fireplace in main living area or Masonry Fireplace in basement
#Fa Fair – Prefabricated Fireplace in basement
#Po Poor – Ben Franklin Stove
#NA No Fireplace

df["FireplaceQu"].unique()
```

```python
#Let’s check the correlation between FireplaceQu and SalePrice, to see how important this feature is in order to determine the price.
#First, we will replace the missing values with NA as per description.
#Then, we encode the categories into numbers from 1 to 5.

df['FireplaceQu'].fillna(0, inplace=True)
df['FireplaceQu'].replace({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5}, inplace=True)
```

```python
#as per barplot we can see FireplaceQu category has an impact on the SalePrice, so we can keep that column
sns.set(style="whitegrid")
sns.barplot(x='FireplaceQu', y="SalePrice", data=df)
```

```python
#lets check other numerical null value column with missing values
#here LotFrontage,MasVnrArea,GarageYrBlt has positive relation with SalePrice but not very strong. so lets drop those columns
columns_with_miss = df.isna().sum()
columns_with_miss = columns_with_miss[columns_with_miss!=0]
c = list(columns_with_miss.index)
c.append('SalePrice')
df[c].corr()
```

```python
df=df.drop(columns=["LotFrontage","MasVnrArea","GarageYrBlt"])
df
```

```python
#in these following column we don't have no  garage or no  basement value as None, so as perr understanding we can convert missing  values as None as well
for column in ['GarageType', 'GarageFinish', 'BsmtFinType2', 'BsmtExposure', 'BsmtFinType1']:
    df[column].fillna('None', inplace=True)

df["MasVnrType"].fillna('None',inplace=True)
df.columns[df.isnull().any()]
```

```python
#we figured out some ordinal data, those have some order and meaning
#we will perform the categorical convertion on the following column and will check the correlation with the SalePrice
cols=['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','GarageQual','GarageCond']
for column in cols:
    df[column].fillna(0,inplace=True)
    df[column].replace({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5}, inplace=True)
cols.append('SalePrice');
df[cols].corr()
```

```python
#most of them have positive correlation with SalePrice, some has not very strong correlation, we will see a clear picture with the bar plot
#we see that those categories have an impact on the sale price, so  we will keep it
f, axes = plt.subplots(2, 4, figsize=(15, 10), sharey=True)

for r in range(0, 2):
    for c in range(0, 4):
        sns.barplot(x=cols.pop(), y="SalePrice", data=df, ax=axes[r][c])

plt.tight_layout()
plt.show()
```

```python
df.columns[df.isnull().any()]
```

```python
#we still have Electrical with missing value and this  column doesn't have prdinal values, so  we will check values count and will replace with most frequent value. we can see  SBrkr coming frequently, so  we  can  replace missing  value with SBrkr
df['Electrical'].value_counts()
```

```python
df['Electrical'].fillna('SBrkr',inplace=True)
```

```python
#now we don't have any column left with missing values
df.columns[df.isnull().any()]
```

```python
#we have devide data into two part to check the result properly, look like nominal categorical data has impact on SalePrice, so we cannot drop it. 
categorical_df = df.select_dtypes(include=['object'])

less_than_equal_six=[]
for column in categorical_df.columns:
    if(len(categorical_df[column].value_counts())<=6):
        less_than_equal_six.append(column)

less_than_equal_six
```

```python
f, axes = plt.subplots(5, 4, figsize=(15, 10), sharey=True)

for r in range(0, 5):
    for c in range(0, 4):
        if less_than_equal_six:
            sns.barplot(x=less_than_equal_six.pop(), y="SalePrice", data=df, ax=axes[r][c])

plt.tight_layout()
plt.show()
```

```python
#we have too much categorical data and can't be 

greater_than_equal_six=[]
for column in categorical_df.columns:
    if(len(categorical_df[column].value_counts())>6):
        greater_than_equal_six.append(column)

greater_than_equal_six
```

```python
f, axes = plt.subplots(6, 2, figsize=(15, 10), sharey=True)

for r in range(0, 6):
    for c in range(0, 2):
        if greater_than_equal_six:
            sns.barplot(x=greater_than_equal_six.pop(), y="SalePrice", data=df, ax=axes[r][c])

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import f_oneway
# Perform one-hot encoding
one_hot_encoded = pd.get_dummies(df['Alley'], prefix='Alley')
# Concatenate the one-hot encoded columns with the original DataFrame
df_encoded = pd.concat([df, one_hot_encoded], axis=1)
# Visualization: Box plot
sns.boxplot(data=df_encoded, x='Alley', y='SalePrice')
```

```python
# ANOVA Test
# print(one_hot_encoded.columns)
one_hot_encoded_columns = one_hot_encoded.columns
category_groups = [df_encoded[df_encoded[column] == 1]['SalePrice'] for column in one_hot_encoded_columns]
f_statistic, p_value = f_oneway(*category_groups)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```
